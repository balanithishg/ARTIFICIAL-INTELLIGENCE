```python
import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt

# Load and prepare data
df = pd.read_excel('super magic(5th).xlsx')
df.columns = [col.strip().lower().replace(" ", "_") for col in df.columns]

# Ensure correct column names
df = df.rename(columns={'date': 'ds', 'oil/day': 'y'})
df['ds'] = pd.to_datetime(df['ds'], errors='coerce')
df['y'] = pd.to_numeric(df['y'], errors='coerce')
df = df.dropna(subset=['ds', 'y'])

# Fit Prophet model
model = Prophet()
model.fit(df)

# Create future dataframe (e.g., next 12 months)
future = model.make_future_dataframe(periods=12, freq='MS')
forecast = model.predict(future)

# Plot forecast
fig1 = model.plot(forecast)
plt.title('Oil Production Forecast')
plt.xlabel('Date')
plt.ylabel('Oil per Day')
plt.tight_layout()
plt.show()

# Optional: forecast components
fig2 = model.plot_components(forecast)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt

# Load data
df = pd.read_excel("super magic(5th).xlsx", sheet_name="oil_production_data_fluctuating")

# Convert Excel serial date to datetime
df['Date'] = pd.to_datetime(df['Date'], origin='1899-12-30', unit='D')

# Filter data between 2008 and 2016
df_filtered = df[(df['Date'] >= '2008-01-01') & (df['Date'] <= '2016-12-31')]

# Prepare for Prophet
df_prophet = df_filtered[['Date', 'oil/day']].rename(columns={'Date': 'ds', 'oil/day': 'y'})

# Initialize and fit Prophet
model = Prophet()
model.fit(df_prophet)

# Make future dataframe within same range
future = df_prophet[['ds']]  # Use same dates for in-sample prediction
forecast = model.predict(future)

# Plot
fig = model.plot(forecast)
plt.title("Prophet Forecast of Oil Production (2008–2016)")
plt.xlabel("Date")
plt.ylabel("Oil/day")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt

# Load data
df = pd.read_excel("super magic(5th).xlsx", sheet_name="oil_production_data_fluctuating")

# Convert Excel serial date to datetime
df['Date'] = pd.to_datetime(df['Date'], origin='1899-12-30', unit='D')

# Filter data between 2008 and 2016
df_filtered = df[(df['Date'] >= '2008-01-01') & (df['Date'] <= '2016-12-31')]

# Prepare for Prophet
df_prophet = df_filtered[['Date', 'oil/day']].rename(columns={'Date': 'ds', 'oil/day': 'y'})

# Train Prophet model
model = Prophet()
model.fit(df_prophet)

# Create future dataframe using same dates
future = df_prophet[['ds']]
forecast = model.predict(future)

# Plot comparison
plt.figure(figsize=(12, 6))
plt.plot(df_prophet['ds'], df_prophet['y'], label='Actual Data', linewidth=2)
plt.plot(forecast['ds'], forecast['yhat'], label='Prophet Prediction', linestyle='--')
plt.fill_between(forecast['ds'], forecast['yhat_lower'], forecast['yhat_upper'], alpha=0.2, label='Confidence Interval')

plt.title('Oil Production Forecast vs Actual (2008–2016)')
plt.xlabel('Date')
plt.ylabel('Oil/day')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
